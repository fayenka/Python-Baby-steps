```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Title')
axes1.set_xlabel('X-Label')
axes1.set_ylabel('Y-Label')

axes2.plot(y,x)
axes2.set_title('Smaller Title')
axes2.set_xlabel('X-Label')
axes2.set_ylabel('Y-Label')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,x**2, label='X')
ax.plot(x,x**3, label='Y')
ax.legend(loc=0)
```
